1. What is a parameter?
    
    -> A parameter is a configuration variable internal to a model that is learned from data during training. For example, in linear regression, the slope and intercept are parameters.

2. What is correlation?
  What does negative correlation mean?

    -> Correlation measures the strength and direction of a relationship between two variables. It ranges from -1 to +1:

    +1: perfect positive correlation

    0: no correlation

    -1: perfect negative correlation

    What does negative correlation mean?

    A negative correlation means that as one variable increases, the other decreases. For example, as the number of hours spent watching TV increases, academic performance might decrease.

3. Define Machine Learning. What are the main components in Machine Learning?

    -> Machine Learning (ML) is a subset of AI that enables systems to learn patterns from data and make predictions or decisions without being explicitly programmed.

    Main components:

    Data: Raw input used for training

    Model: Mathematical structure that learns from data

    Algorithm: Method to train the model (e.g., gradient descent)

    Loss function: Measures prediction error

    Optimizer: Minimizes the loss function

4. How does loss value help in determining whether the model is good or not?

    -> The loss value quantifies how far off a model’s predictions are from actual values. A lower loss indicates better performance. Monitoring loss during training helps detect underfitting or overfitting.

5. What are continuous and categorical variables?

    -> Continuous variables: Numeric values with infinite possibilities (e.g., height, temperature)

    Categorical variables: Represent categories or groups (e.g., gender, color)

6. How do we handle categorical variables in Machine Learning? What are the common techniques?

    -> Since ML models require numerical input, we encode categorical variables:

    Label Encoding: Assigns a unique number to each category

    One-Hot Encoding: Creates binary columns for each category

    Ordinal Encoding: Preserves order in ranked categories

15. What is causation? Explain difference between correlation and causation with an example.

    -> Correlation: Variables move together

    Causation: One variable directly affects another

    Example: Ice cream sales and drowning incidents may correlate (both rise in summer), but one doesn’t cause the other — temperature is the hidden cause.

16. What is an Optimizer? What are different types of optimizers? Explain each with an example.
     
    -> An optimizer updates model parameters to minimize loss.

    Types:

    Gradient Descent: Basic optimizer

    SGD (Stochastic Gradient Descent): Updates per sample

    Adam: Combines momentum and adaptive learning rate

    Example:

    from tensorflow.keras.optimizers import Adam
    model.compile(optimizer=Adam(), loss='mse')
